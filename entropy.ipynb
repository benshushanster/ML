```python
import numpy as np
import matplotlib.pyplot as plt
```

Check the entropy function

```python
def entropy(x):
    p=np.clip(x,a_min=0.00001,a_max=0.99999)
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)

x=np.linspace(0.0, 1.0, 100)
y=entropy(x)
plt.ylabel('entropy')
plt.xlabel('probability')
plt.plot(x,y)
```
